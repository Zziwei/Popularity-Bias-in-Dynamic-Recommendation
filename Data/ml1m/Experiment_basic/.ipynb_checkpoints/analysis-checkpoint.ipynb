```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
from scipy import stats
from tqdm import tqdm

%matplotlib inline
plt.rcParams.update({'font.size': 22})
```

```python
cycle_NDCG_mean = np.load('./cycle_NDCG_mean.npy')
cycle_NDCG_std = np.load('./cycle_NDCG_std.npy')
cycle_click_count_mean = np.load('./cycle_click_count_mean.npy')
cycle_click_count_std = np.load('./cycle_click_count_std.npy')
cycle_GC_click_mean = np.load('./cycle_GC_click_mean.npy')
cycle_GC_click_std = np.load('./cycle_GC_click_std.npy')
cycle_GC_att_mean = np.load('./cycle_GC_att_mean.npy')
cycle_GC_att_std = np.load('./cycle_GC_att_std.npy')
cycle_GC_PR_mean = np.load('./cycle_GC_PR_mean.npy')
cycle_GC_PR_std = np.load('./cycle_GC_PR_std.npy')
itr_item_TPR_mean = np.load('./itr_item_TPR_mean.npy')

truth = np.load('../truth.npy')
audience_size = np.sum(truth, axis=0)
item_sorted = np.argsort(audience_size)

num_cycle = len(cycle_NDCG_mean)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
ax.plot(range(num_cycle), cycle_NDCG_mean, marker='.', linewidth=1.5, markersize=12)
ax.fill_between(range(num_cycle), cycle_NDCG_mean-cycle_NDCG_std, cycle_NDCG_mean+cycle_NDCG_std, alpha=0.5)
ax.set_ylabel('NDCG@20')
ax.set_xlabel('cycle')
ax.set_ylim(0., 0.45)
ax.grid(True)
fig.patch.set_facecolor('white')
# plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
ax.plot(range(num_cycle), cycle_click_count_mean, marker='.', linewidth=1.5, markersize=12)
ax.fill_between(range(num_cycle), cycle_click_count_mean-cycle_click_count_std, cycle_click_count_mean+cycle_click_count_std, alpha=0.5)
ax.set_ylabel('click')
ax.set_xlabel('cycle')
ax.set_ylim(0., 160)
ax.grid(True)
fig.patch.set_facecolor('white')
# plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
ax.plot(range(num_cycle), cycle_GC_click_mean, marker='.', linewidth=1.5, markersize=12)
ax.fill_between(range(num_cycle), cycle_GC_click_mean-cycle_GC_click_std, cycle_GC_click_mean+cycle_GC_click_std, alpha=0.5)
ax.set_ylabel('GC_click')
ax.set_xlabel('cycle')
ax.set_ylim(0., 1.)
ax.grid(True)
fig.patch.set_facecolor('white')
# plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
ax.plot(range(num_cycle), cycle_GC_att_mean, marker='.', linewidth=1.5, markersize=12)
ax.fill_between(range(num_cycle), cycle_GC_att_mean-cycle_GC_att_std, cycle_GC_att_mean+cycle_GC_att_std, alpha=0.5)
ax.set_ylabel('GC_att')
ax.set_xlabel('cycle')
ax.set_ylim(0., 1.)
ax.grid(True)
fig.patch.set_facecolor('white')
# plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
ax.plot(range(num_cycle), cycle_GC_PR_mean, marker='.', linewidth=1.5, markersize=12)
ax.fill_between(range(num_cycle), cycle_GC_PR_mean-cycle_GC_PR_std, cycle_GC_PR_mean+cycle_GC_PR_std, alpha=0.5)
ax.set_ylabel('GC_PR')
ax.set_xlabel('cycle')
ax.set_ylim(0., 1.)
ax.grid(True)
fig.patch.set_facecolor('white')
# plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
''' cumulated true positive rate for items over iterations '''

itr_item_TPR_mean = itr_item_TPR_mean.T
itr_item_TPR_mean = itr_item_TPR_mean[item_sorted, :]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15, 10))
caxes = ax.matshow(itr_item_TPR_mean, aspect='auto')
ax.set_ylabel('items sorted by audience size')
ax.set_xlabel('iteration')
fig.colorbar(caxes) 
fig.patch.set_facecolor('white')
plt.tight_layout()
plt.show()
```
